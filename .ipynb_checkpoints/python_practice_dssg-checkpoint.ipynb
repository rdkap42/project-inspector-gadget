```python
#All columns have spaces, create a function that adds underscore 
import re 

def spaces_to_snake(column_name): 
    s = re.sub(r"\s+", '_', column_name)
    return s.lower()
df.columns = [spaces_to_snake(col) for col in df.columns]

df.columns
```
