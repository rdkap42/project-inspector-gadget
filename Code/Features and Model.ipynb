```python
import pandas as pd
import numpy as py
%pylab inline
```

```python
df = pd.read_csv('/home/benbrew/Documents/Building_Violations_sample_50000.csv')
```

```python
#All columns have spaces, create a function that adds underscore 
import re 

def spaces_to_snake(column_name): 
    s = re.sub(r"\s+", '_', column_name)
    return s.lower()
df.columns = [spaces_to_snake(col) for col in df.columns]

df.columns
```

```python
df.violation_status_date = pd.to_datetime(df.violation_status_date)
df.violation_date = pd.to_datetime(df.violation_date)
```

```python
# look at how many times a location was inspected
inspect = df[['location', 'id']].groupby('location').count()
inspect
```

```python
# look at how many times a location "failed" inspection
failed = df[['location', 'id']][df['inspection_status'] == 'FAILED'].groupby('location').count()
failed
```

```python
dummy=df[['location','violation_code','id']][df['inspection_status']=='FAILED'].groupby(['location','violation_code']).count()
dummy
```

```python
dummy['id']=1
dummy
```

```python
violation_type_count=dummy.reset_index()[['location','id']].groupby('location').count().reset_index()
```

```python
# get unique violation codes for each location
violation_type_count.columns=['locations','unique_violation_codes']
violation_type_count
```

```python
violation_type_count.head(25)
```

```python
violation_type_dist=violation_type_count.copy()
violation_type_dist['locations']=1
violation_type_dist=violation_type_dist.groupby('unique_violation_codes').count().reset_index()
violation_type_dist.columns=['unique_violation_codes','number_of_buildings']
violation_type_dist.head()
```

```python
# Creating dummies
dummies = pd.get_dummies(df[['inspection_status', 'inspection_category']])
new_df = pd.merge(df, dummies, left_index = True, right_index = True) 
new_df.head()
```

```python
# Create a building features dataframe
b_features = pd.DataFrame()
buildings = new_df.groupby('property_group')
```

```python
b_features['num_failed'] = buildings.inspection_status_FAILED.sum()
b_features['num_passed'] = buildings.inspection_status_PASSED.sum()
b_features['total_inspected'] = b_features.num_failed + b_features.num_passed

b_features['num_complaint'] = buildings.inspection_category_COMPLAINT.sum()
```
